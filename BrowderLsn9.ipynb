Create a function that estimates the tip

```python
def est_tip1(total_bill,window):
    x=tips_data[tips_data["total_bill"].between(total_bill-window,total_bill+window,inclusive=True)].mean()
    return x["tip"]
```

```python
est_tip1(20,1)
```

apply this function to the whole data frame and create a new column

```python
tips_data["estimated_tip"]=tips_data["total_bill"].apply(est_tip1,window=(tips_data["total_bill"]*.15))
```

```python
est_tip1(16.99,16.99*.15)
```
